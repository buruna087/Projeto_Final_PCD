## 1. Introdução

As soluções tampão são amplamente utilizadas em laboratório devido à sua resistência a variações drásticas de pH com a adição de pequenas quantidades de um ácido forte ou uma base forte em uma base fraca ou ácido fraco, respectivamente. O efeito tamponante ocorre porque uma solução tampão contém espécies ácidas para neutralizar íons $[\mathrm{OH}^{-}]$ ou espécies básicas para neutralizar os íons $[\mathrm{H}^{+}]$, de modo que essas duas espécies não consumam umas às outras durante a reação. Essas soluções são adquiridas através do processo de titulação, que diz respeito a um reagente adicionado lentamente a uma solução de outro reagente[1].<br>

Em laboratório, as soluções tamponantes são utilizadas de maneiras diversas, como, por exemplo, na prática de PCR (Polymerase Chain Reaction), na qual a presença de uma solução tamponante é indispensável para manter o pH adequado durante as reações de amplificação do DNA.<br>

Com base na teoria e em experiências práticas adquiridas em laboratório, esse projeto foi desenvolvido com o intuito de utilizar a linguagem Python para gerar gráficos da curva de titulação de uma determinada solução, seja ela ácida ou básica, a partir dos dados fornecidos pelo usuário. Essa ferramenta vai possibilitar que sejam observados os comportamentos dessas soluções quando submetidas à titulação, assim, podendo ser de importante uso por estudantes que visam aprender mais sobre o processo ou, até mesmo, observar como a solução preparada em laboratório comporta-se com a ajuda de um gráfico.

### 3.3 Funções 

#### 3.3.1 Funções que calculam o pH

As funções ph_acido_fraco e ph_base_fraca calculam o pH de soluções contendo ácidos ou bases fracos a partir dos valores de pKa ou pKb fornecidos pelo usuário. Inicialmente, esses valores são convertidos em constantes de equilíbrio (Ka ou Kb). Em seguida, é utilizada a equação do equilíbrio químico para obter a concentração de íons $[\mathrm{H}^{+}]$ ou $[\mathrm{OH}^{-}]$ por meio da resolução da equação quadrática correspondente. Por fim, o pH é calculado a partir dessas concentrações. A utilização da solução exata da equação quadrática evita aproximações e aumenta a precisão dos resultados.

```python
#-----------------------------------------------------------
# Função para calcular o pH no modo 1 
#-----------------------------------------------------------

# Para calcular o pH de um ácido fraco, vamos precisar encontrar o x da reação em equlíbrio para encontrar o Ka a partir do pKa recebido. 
def ph_acido_fraco(concentracao, pka): 
    """
    pH de ácido fraco via equação quadrática:
    HA ⇌ H+ + A-   =>   x² + Ka·x - Ka·C = 0
    """
    if concentracao > 0: 
        Ka = 10**(-pka)
        x = (-Ka + np.sqrt(Ka**2+4*Ka*concentracao))/2
        ph = -np.log10(x)
    else:
        raise ValueError("A concentração deve ser maior que zero.")
    return ph 

#-----------------------------------------------------------
# Função para calcular o pH no modo 2 
#-----------------------------------------------------------

# Para uma base fraca vamos trabalhar com o pKb 
def ph_base_fraca(concentracao, pkb):
    """
    pH de base fraca via equação quadrática análoga (usando Kb).
    """
    if concentracao > 0: 
        Kb = 10**(-pkb)
        x = (-Kb + np.sqrt(Kb**2+4*Kb*concentracao))/2
        poh = -np.log10(x)
        ph = 14 - poh 
    else:
        raise ValueError("A concentração deve ser maior que zero.")
    return ph 
```

#### 3.3.2 Funções para calcular a curva de titulação 

Antes de iniciar os cálculos da curvas de titulação, é necessário considerar a escolha feita pelo usuário em relação ás unidades de volume da solução. A função converter_volume é responsável por realizar essa conversão, garantindo a compatibilidade dos dados com as demais funções do programa. Embora o usuário possa fornecer volumes em diferentes unidades, todos os cálculos internos são realizados em litros, mantendo a consistência matemática ao longo da execução. Em seguida, é determinado o volume de equivalência da titulação a partir das concentrações e dos volumes iniciais das soluções.<br>

Com base nesse valor, é criado um conjunto de 500 pontos igualmente espaçados entre volume zero e 160% do volume de equivalência, permitindo representar tanto a região anterior quanto posterior ao ponto de equivalência. Dessa forma, o gráfico não apenas alcança o ponto de equivalência, mas também apresenta a região posterior a ele, permitindo uma visualização mais completa do comportamente do pH ao longo da titulação.<br>

Para cada volume de titulante adicionado, o programa calcula as quantidades de matéria presentes e identifica a região da titulação por meio de estruturas condicionais. Antes da equivalência, o pH é calculado pela equação de Henderson–Hasselbalch, representando a região tampão formada pela coexistência da espécie fraca e de sua forma conjugada. No ponto de equivalência, o pH é determinado considerando a hidrólise da espécie conjugada presente na solução. Após a equivalência, o cálculo passa a ser realizado a partir da concentração do excesso de ácido ou base forte presente no meio.<br>

Em cada iteração, estruturas condicionais(*if*, *elif* e *else*) identificam em qual região da titulação o sistema se encontra e selecionam a equação de pH apropriada para aquele estágio. Os valores calculados são armazenados em uma lista, que posteriormente é utilizada para relacionar o pH ao volume de titulante adicionado e gerar a curva de titulação. Os valores calculados são armazenados em uma lista e posteriormente convertidos em um vetor NumPy, permitindo a construção da curva de titulação por meio da relação entre o pH e o volume de titulante adicionado.

```python
#-----------------------------------------------------------
# Funções para calcular a curva de titulação
#-----------------------------------------------------------

# implementa a conversão de medidas para utilização nas funções da curva de titulação com base na escolha do usuário. 
def converter_volume(V, unidade):
    """ Converte o volume para litros ou mililitros."""
    if unidade == "mL":
        return V / 1000
    elif unidade == "L":
        return V
    else:
        raise ValueError("Unidade deve ser 'mL' ou 'L'")

#-----------------------------------------------------------
# Curva de titulação para o modo 1
#-----------------------------------------------------------

#  Função da curva de titulação para o volume de ácido fraco fixo e adição da base forte 
def curva_titulacao_acido_fraco(
    C_acido, 
    C_base, 
    V_acido,
    unidade,
    pka = None
):
    """
    Curva de titulação de um ácido fraco (fixo) com base forte (adicionada).
 
    Regiões:
      1. moles_base < moles_acido  -> Henderson-Hasselbalch (ou ácido puro se moles_base=0)
      2. moles_base ≈ moles_acido  -> ponto de equivalência (hidrólise de A-)
      3. moles_base > moles_acido  -> excesso de OH- livre
    """
    
    # armazena os valores de pH calculados ao longo da curva 
    ph_lista = []

    # Conversão dos valores 
    V_acido_convertido = converter_volume(V_acido, unidade)

    # Calcula o volume de base necessário para atingir o ponto de equivalência
    V_eq =  (V_acido*C_acido)/C_base

    # cria um vetor com 500 valores de volume de base igualmente espaçados entre 0 e 60% do volume após o ponto de equivalência. 
    V_base = np.linspace(0, 1.6*V_eq,500)

    # Calculo dos moles do ácido(volume fixo)
    moles_acido = C_acido*V_acido_convertido

    for Vb in V_base:
        # Converte o Vb para a unidade escolhida pelo usuário 
        Vb_convertido = converter_volume(Vb, unidade)
       
        # Calcula os moles de base(volume crescente)  
        moles_base = C_base*Vb_convertido

        # Condição para curva quando há mais mols de ácido em comparação aos mols de base em solução
        if moles_base < moles_acido:
            if moles_base == 0: # nenhuma base foi adicionada na solução
                V_total = V_acido_convertido 
                C_solucao = moles_acido/V_total
                pH = ph_acido_fraco(C_solucao, pka)
                ph_lista.append(pH)
            else: # a base começa a ser adicionada
                pH = pka + np.log10(moles_base/(moles_acido - moles_base)) # equação de Henderson-Hasselbalch 
                ph_lista.append(pH)

        # Condição para a curva quando a quantidade de mols de ácido e base são muito próximos 
        elif np.isclose(moles_base, moles_acido,rtol=1e-3): # Se o número for suficientemente próximo de zero, trate-o como zero.
            V_total = V_acido_convertido  + Vb_convertido
            C_sal = moles_acido/V_total # Assume que apenas a hidrólise do ácido conjugado do sal determina o pH
            pKb = 14 - pka 
            pH = ph_base_fraca(C_sal, pKb)
            ph_lista.append(pH)

        # Condição para quando a quantidade de mols de base ultrapassa a quantidade de mols da ácido     
        else:
            V_solucao = V_acido_convertido  + Vb_convertido
            diferenca_moles = moles_base - moles_acido 
            C_OH = np.clip(diferenca_moles/V_solucao, 1e-16, None) # Calcula [OH-] e evita valores inválidos para o cálculo do pOH
            pOH = -np.log10(C_OH)
            pH = 14 - pOH 
            ph_lista.append(pH)

    # Retorna os volumes da base, os valores de pH e o volume de equivalência
    return V_base, np.array(ph_lista), V_eq

#-----------------------------------------------------------
# Curva de titulação para o modo 2
#-----------------------------------------------------------

def curva_titulacao_base_fraca(
    C_acido, 
    C_base,
    V_base,
    unidade,
    pkb= None
):
    """
    Curva de titulação de uma base fraca (fixa) com ácido forte (adicionado).
 
    Regiões:
      1. moles_acido < moles_base  -> Henderson-Hasselbalch para bases (ou base pura se moles_acido=0)
      2. moles_acido ≈ moles_base  -> ponto de equivalência (BH+ formado, ácido fraco)
      3. moles_acido > moles_base  -> excesso de H+ livre
    """
    
    # armazena os valores de pH calculados ao longo da curva 
    ph_lista = []

    # Conversão dos valores 
    V_base_convertido = converter_volume(V_base,unidade)
    
    # Calcula o volume de ácido necessário para atingir o ponto de equivalência
    V_eq =  V_base*C_base/C_acido

    # cria um vetor com 500 valores de volume de ácido igualmente espaçados entre 0 e 60% do volume após o ponto de equivalência. 
    V_acido = np.linspace(0, 1.6*V_eq, 500)

    # Calcula os moles da base(volume fixo)
    moles_base = C_base* V_base_convertido
        

    for Va in V_acido: 
        # Converte o Va para a unidade escolhida pelo usuário 
        Va_convertido = converter_volume(Va, unidade)
        
        # Calcula os moles de ácido(volume crescente) 
        moles_acido = C_acido*Va_convertido

        # Condição para curva quando há mais mols de base em comparação aos mols de ácido em solução 
        if moles_acido < moles_base: 
            if moles_acido == 0: # nenhum ácido foi adicionada na solução
                V_total =  V_base_convertido
                C_solucao = moles_base/V_total
                pH = ph_base_fraca(C_solucao, pkb)
                ph_lista.append(pH)
                
            else: # o ácido começa a ser adicionado
                pOH = pkb + np.log10(moles_acido/(moles_base - moles_acido)) # equação de Henderson-Hasselbalch 
                pH = 14 - pOH
                ph_lista.append(pH)

        # Condição para a curva quando a quantidade de mols de base e ácido são muito próximos 
        elif np.isclose(moles_acido, moles_base,rtol=1e-3): # Se o número for suficientemente próximo de zero, trate-o como zero.
            V_total =  V_base_convertido + Va_convertido 
            C_sal = moles_base/V_total # Assume que apenas a hidrólise da base conjugada do sal determina o pH
            pka = 14 - pkb
            pH = ph_acido_fraco(C_sal, pka)
            ph_lista.append(pH)

        # Condição para quando a quantidade de mols de ácido ultrapassa a quantidade de mols da base 
        else: 
            V_solucao =  V_base_convertido + Va_convertido
            diferenca_moles = moles_acido - moles_base
            C_H = np.clip(diferenca_moles/V_solucao, 1e-16, None) # Calcula [H+] e evita valores nulos ou negativos para o cálculo do pH
            pH = -np.log10(C_H)
            ph_lista.append(pH)

    # Retorna os volumes do ácido, os valores de pH e o volume de equivalência
    return V_acido, np.array(ph_lista), V_eq
```

#### 3.3.3 Funções para plotar o gráfico

Essa célula contém as funções que vão gerar os gráficos a partir da função da curva de titulação definida pelos parâmetros do usuário. Vale ressaltar que, a cada vez que a ferramenta é utilizada, um novo arquivo no formato png é salvo com os novos parâmetros definidos.  

```python
#-----------------------------------------------------------
# plot da curva de titulação com o ácido fraco
#-----------------------------------------------------------

def plotar_curva_acido_fraco(V_base, pH, V_eq, pka,unidade):
    
    plt.plot(V_base, pH)
    
    plt.axvline(V_eq, color='red', linestyle='--', label=f'Ponto de equivalência ({V_eq:.2f} {unidade})')
    
    if pka is not None:
        plt.axvline(V_eq/2, color='green', linestyle=':', label=f'Ponto de tamponamento → pH = pKa = {pka}')
        plt.axhline(pka, color='green', linestyle=':')
        
    plt.xlabel(f'Volume da base ({unidade})')
    
    plt.ylabel("pH")
    
    plt.title("Curva de titulação")
    
    plt.legend()
    
    plt.ylim(0, 14)
    
    plt.grid(True, alpha=0.3)

    plt.savefig("Curva_de_Titulacão_modo1.png", dpi=150)

    plt.show()

    # Salva diferentes imagens para cada vez que a ferramenta é utilizada
    plt.close()
    
#-----------------------------------------------------------
# plot da cruva de titulação com a base fraca
#-----------------------------------------------------------

def plotar_curva_base_fraca(V_acido, pH, V_eq, pka, unidade):
    
    plt.plot(V_acido, pH)
    
    plt.axvline(V_eq, color='red', linestyle='--', label=f'Ponto de equivalência ({V_eq:.2f} {unidade})')
    
    if pka is not None:
        # no ponto de tamponamento metade da espécie fraca original foi neutralizada e convertida em sua espécie conjugada.
        plt.axvline(V_eq/2, color='green', linestyle=':', label=f'Ponto de tamponamento → pH = pKa = {pka}')
        plt.axhline(pka, color='green', linestyle=':')
        
    plt.xlabel(f'Volume do ácido ({unidade})')
    
    plt.ylabel("pH")
    
    plt.title("Curva de titulação")
    
    plt.legend()
    
    plt.ylim(0, 14)
    
    plt.grid(True, alpha=0.3)

    plt.savefig("Curva_de_Titulacão_modo2.png", dpi=150)

    plt.show()
    
    # Salva diferentes imagens para cada vez que a ferramenta é utilizada
    plt.close()
```

Em síntese, a união entre os conhecimentos na linguagem Python, as observações das práticas ministradas em laboratório e a fundamentação teórica adquirida no decorrer do semestre letivo permitiram que este projeto fosse executado de maneira interdisciplinar, atendendo ao objetivo do trabalho final da disciplina de Prática em Ciência de Dados.
